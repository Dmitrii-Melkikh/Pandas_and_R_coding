# Задание 4

1. Для категорий из задания 2 найдите среднюю и медианную цену по категории в пределах каждого месяца.
2. В каждом году найдите месяцы, в которых средняя цена меньше, чем в предыдущем и следующем месяце.

```python

def opr_month(l):
  months=['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
  l1=[]
  l2=[]
  l3=[]
  for i in l:
    # print(int(i)//12)
    if int(i)//12==0:
      l1.append(int(i)%12)
    elif int(i)//12==1:
      l2.append(int(i)%12)
    else:
      l3.append(int(i)%12)
  for i in range(len(l1)):
    l1[int(i)]=months[l1[int(i)]]
  for i in range(len(l2)):
    l2[int(i)]=months[l2[int(i)]]
  for i in range(len(l3)):
    l3[i]=months[l3[i]]
  return l1,l2,l3
  
  
  



import pandas as pd
id_max=40
id_min=10
sales_train=pd.read_csv("/content/drive/MyDrive/data/sales_train.csv")
items=pd.read_csv("/content/drive/MyDrive/data/items.csv")
item_categories=pd.read_csv("/content/drive/MyDrive/data/item_categories.csv")
df=sales_train.merge(items)
months_mean_max=[0]*34
months_median_max=[0]*34
months_mean_min=[0]*34
months_median_min=[0]*34
for i in range(34):
  df_1=df[(df['category_id']==id_max) & (df['date_block_num']==i)]
  df_2=df[(df['category_id']==id_min) & (df['date_block_num']==i)]
  months_mean_max[i]=df_1['item_price'].mean()
  months_median_max[i]=df_1['item_price'].median()
  months_mean_min[i]=df_2['item_price'].mean()
  months_median_min[i]=df_2['item_price'].median()
p_max=pd.DataFrame()
p_max['month']=range(34)
p_max['mean_price']=months_mean_max
p_max['median_price']=months_median_max

p_min=pd.DataFrame()
p_min['month']=range(34)
p_min['mean_price']=months_mean_min
p_min['median_price']=months_median_min

print('Для категории с максимальными продажами  (' + str(id_max) + ')  :')
print(p_max)
print('Для категории с минимальными продажами  (' + str(id_min) + ')  :')
print(p_min)


less_month_max=[]
less_month_min=[]
for i in range(len(p_max)):
  if i!=0 and i!=33:
    if p_max.iloc[i-1]['mean_price']>p_max.iloc[i]['mean_price'] and p_max.iloc[i]['mean_price']<p_max.iloc[i+1]['mean_price']:
      less_month_max.append(p_max.iloc[i]['month'])
    if p_min.iloc[i-1]['mean_price']>p_min.iloc[i]['mean_price'] and p_min.iloc[i]['mean_price']<p_min.iloc[i+1]['mean_price']:
      less_month_min.append(p_min.iloc[i]['month'])



less_month_max_2013, less_month_max_2014, less_month_max_2015=opr_month(less_month_max)
less_month_min_2013, less_month_min_2014, less_month_min_2015=opr_month(less_month_min)
print('Месяцы, в которых средняя цена меньше, чем в предыдущем и следующем месяце для максимальной категории (40):')
print('2013:')
print(less_month_max_2013)
print('2014:')
print(less_month_max_2014)
print('2015:')
print(less_month_max_2015)


print('Месяцы, в которых средняя цена меньше, чем в предыдущем и следующем месяце для минимальной категории (10):')
print('2013:')
print(less_month_min_2013)
print('2014:')
print(less_month_min_2014)
print('2015:')
print(less_month_min_2015)










```
